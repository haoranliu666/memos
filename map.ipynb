```python
import folium
import pandas as pd
import random

df = pd.read_stata('/Users/haoranliu/GRA/portfolio_did/temp.dta')

locations = df[['latitude_y', 'longitude_y']]
location_list = locations.values.tolist()
sampled_locations = random.sample(location_list, k=100)

# Create map centered on the United States
us_map = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

# Add markers for each location
for point in range(0, len(sampled_locations)):
    folium.Marker(sampled_locations[point]).add_to(us_map)
```
